```python
%load_ext autoreload
%autoreload 2

from src.game import Game
from src.agent import Random_Agent, QAgent
from matplotlib import pyplot as plt
import numpy as np
```

# Training Notebook

## Train the agent using q-learning


```python
env = Game(randomize_start_pos=False)
agent = QAgent()
agent.train(env, episodes=5000, epsilon=1, min_epsilon=0.001, print_every=500,epsilon_decay=.995)
```

Plot the rewards obtained over the course of training.

```python
agent.plot_train_stats(path=None)
plt.hlines(1250,xmin=0,xmax=5000, colors='gold')
plt.annotate('Convergence',(0,1300))
```

The agent started to converge around 3000 episodes for this example.  The time of convergence may vary when you run this multiple times, but convergence typically occurs around 2000-4000 episodes in most cases.

## Compare the performance of the Q agent versus a Random baseline

```python
## Performance of Q agent
q_agent_rewards = agent.performance(env)
random_agent_rewards = np.mean([Random_Agent(5,4).performance(env) for i in range(100)])

print(f"Average rewards for random agent: {random_agent_rewards}\nRewards obtained by Q-agent: {q_agent_rewards}")
```

Our agent recieved the maximum score and beat just a randomly walking agent!
